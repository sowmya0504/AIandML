```python
import pandas as pd
import numpy as np
import sklearn
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/Mounika-Kajjam/Datasets/master/weather_data.csv",parse_dates=["day"])
data
```

```python
data=data.replace({'temperature':-99999,"windspeed":-99999,"event":'0'},np.nan)
data
```

```python
### interpolate() it adds random values in the range 

int_data=data.interpolate()
int_data
```

```python
### dropna()
drop_data=data.dropna()
drop_data
```

```python
### dropna()
drop_data=data.dropna(how="any")
drop_data
```

```python
### dropna()
drop_data=data.dropna(how="any",axis=1)
drop_data
```

```python
# how = all drops the rows or columns that have all the values as nan
drop_data1=data.dropna(how="all")
drop_data1
```

### handling the missing vlaues using scikit-learn

```python
### Simple imputer, it will only work for numerical columns
# creating an object for SimpleImputer class
from sklearn.impute import SimpleImputer
imputer =SimpleImputer(missing_values=np.nan,strategy="mean")
```

```python
# it is a 2 step transformation
# applying the fit method on the data
imputer.fit(data.iloc[:,1:3])#indices of column 1 and 2 
# we gave the index as 3 it is excluded
```

```python
# transform the data
data.iloc[:,1:3]=imputer.transform(data.iloc[:,1:3])
data
```

### Encoding the Categorical imputer

### categorical data types
- nominal
    - has no numerical hierarchy
    - no relation ship between the categories
- ordinal
    - has a level of hierarchy
    - relation ship between the categories exists

```python
# usiong pandas --> pd.get_dummies()
price=pd.read_csv('https://raw.githubusercontent.com/Mounika-Kajjam/Datasets/master/homeprices.csv')
price
```

```python
# convert categorical data to numerical columns
dummy_set=pd.get_dummies(price.town)
dummy_set
# 1 represents the presence of the category and 0 represents the absence
```

```python
# concatinate the 2 data sets the dummy_set and price
merged_data=pd.concat([price,dummy_set],axis=1)
merged_data
```

```python
merged_data=pd.concat([price,dummy_set])
merged_data
```

```python
pd.get_dummies(price,columns=['town'])
```

```python
# when we add n dummy colummns then we need to have n-1 dummy columns
pd.get_dummies(price,columns=['town'],drop_first=True)
```
