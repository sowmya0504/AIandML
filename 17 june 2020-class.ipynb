```python
### OneHotEncoder --> similar to pd.get_dummies
    -- import onehotencoder
    -- create an object for the class
    -- column transformer --> inside the columntransformer,we will pass onehotencoder
    -- it has 2 mandatory arguments --> transforms,remainder
    -- transformers will accept a list of tupple values
```

### Feature scaling (Not Mandatory for every algorithms)


```python
sample=pd.DataFrame({"w.e":[2,4,5,6,7,8],"age":[28,25,32,45,33,32],"salary":[300000,400000,500000,600000,700000,800000],"distance":[10,12,23,32,15,17],"expected salary":[400000,500000,600000,700000,800000,900000]})
sample
```

-- most of the machine learning algoriths ,we use distance based formulas
-- all the columns in the data set belongs to different uints
-- units and scale(magnnitude)
-- feature scaling --> when applied units and scaling will be removied
-- to make the data unitless and scaleless

```python
from sklearn.preprocessing import scale
scaled_data=scale(sample)
scaled_data
```

```python
scaled_data=pd.DataFrame(scaled_data)
scaled_data
```

```python
standard_data=sample.copy() #creating copy of original data
standard_data
```

```python
## using standardscaler
from sklearn.preprocessing import StandardScaler
sc=StandardScaler() #creating a object 
sample = sc.fit_transform(sample)
sample
```
