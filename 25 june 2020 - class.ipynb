```python
#importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## loading the boston dataset
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
boston_df= pd.DataFrame(boston.data,columns=boston.feature_names)
boston_df.head()
```

```python
boston_df['MEDV'] = boston.target
boston_df.head()
```

```python
## split the data into training and testing;Build the model on training data and test the model on testing data
```

```python
# split the data into Input and Output
X=boston_df.drop('MEDV',axis=1) #input columns
y=boston_df.MEDV #boson_df['MEDV']
```

```python
# Divide the X and y into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
#Build the model on training data --> X_trin and y_train
# sklearn library:import,instantiate,fit
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
# Intercept and the coefficient values
print(lm.intercept_)
lm.coef_
```

```python
## Create a dataframe for coefficientss
coefficients = pd.DataFrame([X_train.columns])
coefficients
```

```python
## Test the model on testing data
y_test_pred = lm.predict(X_test) # test data --> unseen data
y_test_pred
```
